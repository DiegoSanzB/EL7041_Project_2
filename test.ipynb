## Just modulated

### QPSK

```python
data = pd.read_csv('results/QPSK.csv')
data = data[ ["re", "im"]]
data_array = data.to_numpy()

color_dots = []
for i in range(len(data_array)):
    color_dots.append(COLOR[str(MOD4.index(data_array[i].tolist()))])
# Separate coordinates
real = data_array[:, 0]
imag = data_array[:, 1]
# Create a figure and a set of subplots
fig1, axs1 = plt.subplots(1)
axs1.set_title("Modulated data")
axs1.set_xlabel(r"$\Re$(-)")
axs1.set_ylabel(r"$\Im$(-)")
axs1.set_xlim(-1.70, 1.7)  # Set limits for the x-axis
axs1.set_ylim(-1.5, 1.5)  # Set limits for the y-axis
# axs1.grid()
# Plot the data
axs1.scatter(real, imag, color = color_dots, zorder = 7)
# Insert handmade legend
axs1.legend(handles = LEG4)
# axs1.figure.savefig('results/clean_QPSK.pdf')
```

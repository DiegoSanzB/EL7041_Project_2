# Plot modulated data

## Just modulated

### QPSK

```python
data = pd.read_csv('results/QPSK.csv')
data = data[ ["re", "im"]]
data_array = data.to_numpy()
# De-normalize for colors
data_colors = np.rint(data_array*np.sqrt(2))
color_dots4 = []
for i in range(len(data_array)):
    color_dots4.append(COLOR[str(MOD4.index(data_colors[i].tolist()))])
# Separate coordinates
real = data_array[:, 0]
imag = data_array[:, 1]
# Create a figure and a set of subplots
fig1, axs1 = plt.subplots(1)
axs1.set_title("Modulated data")
axs1.set_xlabel(r"$\Re$(-)")
axs1.set_ylabel(r"$\Im$(-)")
axs1.set_xlim(-1.70, 1.7)  # Set limits for the x-axis
axs1.set_ylim(-1.5, 1.5)  # Set limits for the y-axis
# axs1.grid()
# Plot the data
axs1.scatter(real, imag, color = color_dots4, zorder = 7)
# Insert handmade legend
axs1.legend(handles = LEG4)
# axs1.figure.savefig('results/clean_QPSK.pdf')
```

## With noise

### QPSK

```python
data = pd.read_csv('results/QPSK_snr.csv')
# create a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2)
fig.suptitle("QPSK_SNR")
# Get the SNR values
# Iterate over the subplots
for i in range(4):
    # Select the subplot
    ax = axs[i // 2, i % 2]
    ax.set_title("SNR = " + str(SNR[i]) + " dB")
    ax.set_xlabel(r"$\Re$(-)")
    ax.set_ylabel(r"$\Im$(-)")
    # ax.set_xlim(-1.70, 1.7)  # Set limits for the x-axis
    # ax.set_ylim(-1.5, 1.5)  # Set limits for the y-axis
    # Separate coordinates and get them as a numpy array
    real = data[f"re_noise_{SNR[i]}"].to_numpy()
    imag = data[f"im_noise_{SNR[i]}"].to_numpy()
    # Plot the data
    ax.scatter(real, imag, color = color_dots4, zorder = 7)
    # Insert handmade legend
# Set legend outside the plot
fig.legend(handles=LEG4, bbox_to_anchor=(0.9, 0.5), loc='center left')

# Adjust layout to prevent overlap with the title
plt.tight_layout(rect=[0, 0, 0.9, 0.95])
# print("Done")
# plt.savefig('results/QPSK_SNR.png')
print("Done")
plt.show()
```

# Plot modulated data

## Just modulated

### QPSK

```python
data = pd.read_csv('results/QPSK.csv')
data = data[ ["clean_re", "clean_im"]]
data_array = data.to_numpy()
# De-normalize for colors
data_colors = np.rint(data_array*np.sqrt(2))
color_dots4 = []
for i in range(len(data_array)):
    color_dots4.append(COLOR[str(MOD4.index(data_colors[i].tolist()))])
# Separate coordinates
real = data_array[:, 0]
imag = data_array[:, 1]
# Create a figure and a set of subplots
fig1, axs1 = plt.subplots(1)
axs1.set_title("Modulated data")
axs1.set_xlabel(r"$\Re$(-)")
axs1.set_ylabel(r"$\Im$(-)")
axs1.set_xlim(-1.70, 1.7)  # Set limits for the x-axis
axs1.set_ylim(-1.5, 1.5)  # Set limits for the y-axis
# axs1.grid()
# Plot the data
axs1.scatter(real, imag, color = color_dots4, zorder = 7, s=s)
# Insert handmade legend
axs1.legend(handles = LEG4)
axs1.figure.savefig('results/clean_QPSK.png')
```

### 16QAM

```python
data = pd.read_csv('results/16QAM.csv')

data = data[ ["clean_re", "clean_im"]]
data_array = data.to_numpy()
data_colors = np.rint(data_array*np.sqrt(10))
color_dots16 = []
for i in range(len(data_array)):
    color_dots16.append(COLOR[str(MOD16.index(data_colors[i].tolist()))])
# Separate coordinates
real = data_array[:, 0]
imag = data_array[:, 1]
# Create a figure and a set of subplots
fig1, axs1 = plt.subplots(1)
axs1.set_title("Modulated data")
axs1.set_xlabel(r"$\Re$(-)")
axs1.set_ylabel(r"$\Im$(-)")
axs1.set_xlim(-1.20, 1.9)  # Set limits for the x-axis
axs1.set_ylim(-1.20, 1.3)  # Set limits for the y-axis
# axs1.grid()
# Plot the data
axs1.scatter(real, imag, color = color_dots16, zorder = 7, s=s)
# Insert handmade legend
axs1.legend(handles=LEG16)
axs1.figure.savefig('results/clean_16QAM.png')
```

```python
data = pd.read_csv('results/QPSK_eq_doppler_perfect.csv')
# create a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2)
fig.suptitle("QPSK-Doppler Perfect Channel Estimation")
# Get the SNR values
# Iterate over the subplots
for i in range(4):
    # Select the subplot
    ax = axs[i // 2, i % 2]
    ax.set_title("SNR = " + str(SNR[i]) + " dB")
    ax.set_xlabel(r"$\Re$(-)")
    ax.set_ylabel(r"$\Im$(-)")
    # ax.set_xlim(-2.5, 2.5)  # Set limits for the x-axis
    # ax.set_ylim(-2.5, 2.5)  # Set limits for the y-axis
    # Separate coordinates and get them as a numpy array
    real = data[f"{SNR[i]}_re"].to_numpy()
    imag = data[f"{SNR[i]}_im"].to_numpy()
    
    if i in [0, 1, 2]:
        x_max = np.percentile(real, 99.9)
        y_max = np.percentile(imag, 99.9)
        ax.set_xlim(-x_max, x_max)
        ax.set_ylim(-y_max, y_max)
        print(x_max, y_max)
    # Plot the data
    ax.scatter(real, imag, zorder = 7, s=s)
# Set legend outside the plot
# fig.legend(handles=LEG16, bbox_to_anchor=(0.9, 0.5), loc='center left')
# Adjust layout to prevent overlap with the title
plt.tight_layout(rect=[0, 0, 0.9, 0.95])
# print("Done")
plt.savefig('results/QPSK_DOPPLER_PERFECT.png', bbox_inches='tight')
print("Done")
plt.show()
```
